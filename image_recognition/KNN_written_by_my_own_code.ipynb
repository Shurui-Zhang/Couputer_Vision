```python
def calculate_distance(image_test, image_train):
  distance = 0
  for i in range(256):
    distance += (image_test[0][i] - image_train[0][i]) ** 2
  
  return distance ** 0.5
```

```python
def sort_distances(train_loader, image_test):
  distances = []
  new_image_test = image_test.view(1,256)
  for image_train, label_image_train in train_loader:  
    new_image_train = image_train.view(1,256)
    dist = calculate_distance(new_image_test, new_image_train) 
    distances.append((label_image_train, dist))
    distances.sort(key=operator.itemgetter(1))
  return distances
```

```python
def get_prediction_class(distances, k):
  neighbour = []
  for i in range(k):
    neighbour.append(distances[i][0])
  classname_times = {}

  for classname in neighbour:
    if classname in classname_times:
      classname_times[classname] += 1
    else:
      classname_times[classname] = 1
  new_class_times = sorted(classname_times.items(), key=operator.itemgetter(1), reverse=True)
  prediction_class = new_class_times[0][0]
  return prediction_class
```

```python
def knn(k, train_loader, test_loader, images, classes_name):
  i = 0
  image_prediction = []
  for image_test, _ in test_loader:
    distances = sort_distances(train_loader, image_test)
    prediction = get_prediction_class(distances, k)
    prediction_class = classes_name[prediction]
    image_prediction.append((images[i][0], prediction_class))
    i += 1
  return image_prediction
```
